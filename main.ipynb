# Parte 1 – Filtrado y preprocesamiento

```python
# Importar las librerías necesarias
import pandas as pd
import json
```

```python
# 2. Cargar la información del archivo large_even.json en una lista
with open('large_eve.json', 'r') as f:
    data = [json.loads(line) for line in f]
print(f"Total de registros: {len(data)}")  # Debería ser 746909
```

```python
# 3. Filtrar registros DNS
data_dns = [record for record in data if record.get('event_type') == 'dns']
```

```python
# 4. Mostrar la nueva cantidad de registros filtrados
print(f"Total de registros DNS: {len(data_dns)}")  # Debería ser 21484
```

```python
# 5. Mostrar la información de 2 registros cualesquiera
print(data_dns[0])
print(data_dns[100])
```

```python
# 6. Normalizar la información y asignarla en un dataframe
df = pd.json_normalize(data_dns)
print(df.shape)  # Debería ser (21484, 163)
```

```python
df
```

```python
# 7. Filtrar los registros DNS para aquellos registros tipo A
df_a = df[df['dns.rrtype'] == 'A'].copy()
print(df_a.shape)  # Debería ser (2849, 163)
df_a.head()
```

```python
# 8. Filtrar los dominios únicos
unique_domains = df_a['dns.rrname'].unique()
print(len(unique_domains))  # Debería ser 177
```

```python
df_a['dns.rrname']
```

```python
# 9. Escribir una función que obtenga el TLD para un dominio
def get_tld(domain):
    # Si el dominio termina con '.home', entonces 'home' es el TLD
    if domain.endswith('.home'):
        return 'home'
    # De lo contrario, el TLD es el segundo nivel de dominio y el TLD
    else:
        parts = domain.split('.')
        tld = '.'.join(parts[-2:])
        return tld
```

```python
# 10. Del dataframe de dominios únicos de tipo A, obtener el TLD
df_a['domain_tld'] = df_a['dns.rrname'].apply(get_tld)
df_a = df_a[['domain_tld']]  # Eliminar todas las demás columnas
```

# Parte 3 – Dominio experto

```python
# Cargar la lista de TLDs
with open('top-1m.csv', 'r') as f:
    tld_list = [line.strip().split(',')[1] for line in f]

# Convertir la lista a un conjunto para búsquedas más rápidas
tld_set = set(tld_list)

def is_in_tld_list(tld):
    return 0 if tld in tld_set else 1
```

```python
df_dga_unique['not_in_tld_list'] = df_dga_unique['domain_tld'].apply(is_in_tld_list)
df_suspicious = df_dga_unique[df_dga_unique['not_in_tld_list'] == 1]
```

```python
import whois

def get_creation_date(domain):
    try:
        w = whois.whois(domain)
        return w.creation_date
    except Exception as e:
        print(f"Error obteniendo la fecha de creación de {domain}: {e}")
        return None
```

```python
df_suspicious['creation_date'] = df_suspicious['domain_tld'].apply(get_creation_date)
```
